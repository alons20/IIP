```python
#lista con numeros
#Orozco Jose

numeros = [45, 12, 17, 78, 3, 56, 89, 23, 67, 1, 34]
print('Origuinal:', numeros)

#Ordenar de menor a mayor
numeros.sort()
print('Ordena:', numeros)

#Ordena de mayor a menor
numeros.reverse()
print('Invertida:', numeros)
```

```python
#Promedio
#Orozco Jose

numeros = [45, 12, 17, 78, 3, 56, 89, 23, 67, 1, 34]

total = sum(numeros)
promedio = sum(numeros) / len(numeros)
print(f'Suma: {total}')
print(f'Promedio: {promedio:.2f}')
```
